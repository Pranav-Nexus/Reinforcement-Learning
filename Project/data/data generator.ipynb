```python
import random
from datetime import datetime, timedelta

start_date = datetime.strptime("2016-12-30", "%Y-%m-%d")
end_date = datetime.strptime("2024-03-31", "%Y-%m-%d")
delta = timedelta(days=1)

current_date = start_date
while current_date <= end_date:
    demand = float(random.randint(90000, 397000))
    supply = float(random.randint(31300, 300000))
    print(f"{current_date.strftime('%Y-%m-%d')},{demand:.1f},{supply:.1f}")
    current_date += delta
```
